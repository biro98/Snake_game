```python
#run Set_up for windows initialization
%run 'Set_up.ipynb'

#run Score_and_game_over 
%run 'Score_game_over_speed.ipynb'

class Main:
    def __init__(self, snake_game):
        self.snake_game = snake_game

    def run(self):
        while True:
            # handling key events and movement
            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_UP:
                        self.snake_game.change_to = 'UP'
                    if event.key == pygame.K_DOWN:
                        self.snake_game.change_to = 'DOWN'
                    if event.key == pygame.K_LEFT:
                        self.snake_game.change_to = 'LEFT'
                    if event.key == pygame.K_RIGHT:
                        self.snake_game.change_to = 'RIGHT'

            # assuring the direction of the snake towards one direction only
            if self.snake_game.change_to == 'UP' and self.snake_game.direction != 'DOWN':
                self.snake_game.direction = 'UP'
            if self.snake_game.change_to == 'DOWN' and self.snake_game.direction != 'UP':
                self.snake_game.direction = 'DOWN'
            if self.snake_game.change_to == 'LEFT' and self.snake_game.direction != 'RIGHT':
                self.snake_game.direction = 'LEFT'
            if self.snake_game.change_to == 'RIGHT' and self.snake_game.direction != 'LEFT':
                self.snake_game.direction = 'RIGHT'

            # Moving the snake
            if self.snake_game.direction == 'UP':
                self.snake_game.snake_position[1] -= 10
            if self.snake_game.direction == 'DOWN':
                self.snake_game.snake_position[1] += 10
            if self.snake_game.direction == 'LEFT':
                self.snake_game.snake_position[0] -= 10
            if self.snake_game.direction == 'RIGHT':
                self.snake_game.snake_position[0] += 10

           
            
            #Update and handle obstacles
            self.snake_game.update_obstacles()
            for obstacle in self.snake_game.obstacles:
                if obstacle.moving:
                    obstacle.move()
                    obstacle.check_screen_edges(self.snake_game.window_x, self.snake_game.window_y)
                obstacle.draw(self.snake_game.game_window, self.snake_game.red)
                # Calculate the distance between the center of the snake's head and the obstacle
                distance = ((self.snake_game.snake_position[0] - obstacle.position[0])**2 + 
                            (self.snake_game.snake_position[1] - obstacle.position[1])**2)**0.5
                if distance < (10/2 + obstacle.radius):  # Assuming snake block size is 10x10
                    SnakeGameUtilities.game_over(self.snake_game.game_window, self.snake_game.window_x, self.snake_game.window_y, self.snake_game.score, self.snake_game.red)
                

            # Snake body growing mechanism
            self.snake_game.snake_body.insert(0, list(self.snake_game.snake_position))
            if self.snake_game.snake_position[0] == self.snake_game.fruit_position[0] and self.snake_game.snake_position[1] == self.snake_game.fruit_position[1]:
                self.snake_game.score += 10
                self.snake_game.fruit_spawn = False

                # Correctly calling the increase_speed method
                self.snake_game.snake_speed = SnakeGameUtilities.increase_speed(self.snake_game.score, self.snake_game.snake_speed)
            else:
                self.snake_game.snake_body.pop()

            if not self.snake_game.fruit_spawn:
                self.snake_game.fruit_position = [random.randrange(1, (self.snake_game.window_x//10)) * 10,
                                                  random.randrange(1, (self.snake_game.window_y//10)) * 10]

            self.snake_game.fruit_spawn = True
            
            # Drawing the game elements
            self.snake_game.game_window.fill(self.snake_game.black)
            
            #Draw snake, fruit, and obstacles
            for pos in self.snake_game.snake_body:
                pygame.draw.rect(self.snake_game.game_window, self.snake_game.green, pygame.Rect(pos[0], pos[1], 10, 10))

            pygame.draw.rect(self.snake_game.game_window, self.snake_game.white, pygame.Rect(self.snake_game.fruit_position[0], self.snake_game.fruit_position[1], 10, 10))

            for obstacle in self.snake_game.obstacles:
                obstacle.draw(self.snake_game.game_window, self.snake_game.red)  # Draw each obstacle


            # Game Over conditions
            if self.snake_game.snake_position[0] < 0 or self.snake_game.snake_position[0] > self.snake_game.window_x-10:
                SnakeGameUtilities.game_over(self.snake_game.game_window, self.snake_game.window_x, self.snake_game.window_y, self.snake_game.score, self.snake_game.red)
            if self.snake_game.snake_position[1] < 0 or self.snake_game.snake_position[1] > self.snake_game.window_y-10:
                SnakeGameUtilities.game_over(self.snake_game.game_window, self.snake_game.window_x, self.snake_game.window_y, self.snake_game.score, self.snake_game.red)

            # Touching the snake body
            for block in self.snake_game.snake_body[1:]:
                if self.snake_game.snake_position[0] == block[0] and self.snake_game.snake_position[1] == block[1]:
                    self.snake_game.game_over()

            
            # Displaying score continuously
            SnakeGameUtilities.show_score(self.snake_game.game_window, self.snake_game.score, 1, self.snake_game.white, 'times new roman', 20)

            # Refresh game screen
            pygame.display.update()

            # Frame Per Second /Refresh Rate
            self.snake_game.fps.tick(int(self.snake_game.snake_speed))
            
            
#run the Game           
if __name__ == "__main__":
    game = SnakeGame()
    main = Main(game)
    main.run()
```
